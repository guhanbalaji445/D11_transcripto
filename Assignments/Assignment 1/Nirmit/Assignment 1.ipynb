```python
#TASK 1
import pandas as pd
data = pd.read_csv('vgchartz-2024.csv') # reads the csv file, loads the csv
#file into a dataframe 
data.head() #prints the first 5 rows of the csv file

```

```python
data.info() # gives overall info of the csv file
```

```python
data.describe() # gives the statistics on the data(only int and float data types)
```

```python
data.isnull()
# True represents the missing values
```

```python
data['missing in critic_score'] = data['critic_score'].isnull()
#Creates a new column which flags(gives info) whether the value is missing or not in any 
#given row, on the specified column I'm working on 
data.head()
```

```python
import numpy as np
#Creating a new column 'total_sales'
data['total_sales']=data['na_sales']+data['jp_sales']+data['pal_sales']+data['other_sales']
mean_val=np.mean(data['total_sales'])
med_val=np.median(data['total_sales'])
std_val=np.std(data['total_sales'])
mean_val2=np.mean(data['critic_score'])
print('Mean= ',mean_val)
print('Median= ',med_val)
print('Standard Deviation= ',std_val)
print(mean_val2)
```

```python
#Filtering the rows where total_sales is greater than 13
print(data[data['total_sales']>13])
```

```python
print(data[['critic_score','other_sales']])

#Vectorized functions of a numpy array makes the code faster and optimizes computations.
#It is better than lists of python as in a list we have run a loop to update each value,
#whereas in an array of numpy, the operations are done on the entire array at the same time.

#We are using these vectorized functions to update the values of 'critic_scores' and 'other_sales'
data['critic_score']=data['critic_score']-0.1
data['other_sales']=data['other_sales']+1.5
print(data[['critic_score','other_sales']])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns #seaborn is a python visualization library built on the roof of Matplotlib. It is used to generate
#data visualizations which are more aesthetic.
sns.histplot(data['total_sales'], bins=20,kde=True, color='blue',edgecolor='black')
plt.title('Distribution of Total Sales')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend(['Density Plot','Distribution'],loc="upper right") 
plt.show()
```

```python
plt.scatter(data['critic_score'],data['total_sales'],alpha=0.6,marker='H',color='lightgreen')
plt.title('Critic Score vs Total Sales')
plt.xlabel('Critic Score')
plt.ylabel('Total Sales')
plt.show()
#From the given scatter plot we can infer that the total sales increase as the crtic score increases.
#Thus people tend to buy games which has a higher critic score
```
